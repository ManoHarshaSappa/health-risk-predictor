```python
print("Feature"+" "*25+"null val count"+" "*5+"null val percentage" )
print("-"*70)
for feature in df:
    total_missing_values= df[feature].isnull().sum()
    percentage_missing_values= total_missing_values/25000
    if total_missing_values>0:
        print(feature," "*(30-len(feature)), total_missing_values," "*(20-len(str(total_missing_values))),f'{percentage_missing_values*100:.2f} %')
```
